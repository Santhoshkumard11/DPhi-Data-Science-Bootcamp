# Summary: Data Preprocessing

```python
import pandas as pd
import numpy as np
```

```python
titanic = pd.read_csv('train - train_titanic.csv')
titanic1 = pd.read_csv('train - train_titanic.csv')
titanic.head()
```

## Drop Multiple Columns

```python
titanic  = titanic.drop(['PassengerId','Name', 'Ticket'], axis = 1)
titanic.head()
```

## Creating Dummy Variables and Concatenating the DataFrames

```python
dummies = []

cols = ['Pclass', 'Sex', 'Embarked']

for col in cols:
    dummies.append(pd.get_dummies(titanic1[col]))
    
titanic_dummies = pd.concat(dummies, axis=1) 
titanic_dummies.head()
```

```python
train_data1 = pd.concat((titanic,titanic_dummies), axis=1)
train_data1.head()
```

## Rename Column Names

```python
train_data = train_data1.copy()
train_data.rename(columns={1: 'Pclass_1', 2: 'Pclass_2', 3: 'Pclass_3', 'C': 'Emb_C', 'Q': 'Emb_Q', 'S': 'Emb_S' }, inplace = True)
train_data = train_data.drop(['Pclass', 'Sex', 'Embarked'], axis = 1)
```
